```python
from __future__ import division, print_function, unicode_literals
import numpy as np
import matplotlib.pyplot as plt
# Temp (F degree)
X = np.array([[194.5, 194.3, 197.9, 198.4, 199.4, 199.9, 200.9, 201.1, 201.4, 201.3, 203.6, 204.6, 209.5, 208.6, 210.7, 211.9, 212.2]]).T
# Press (Atm)
y = np.array([[20.79,20.79,22.4,22.67,23.15,23.35,23.89,23.99,24.02,24.01,25.14,26.57,28.49,27.76,29.04,29.88,30.06]]).T
```

```python
# Visual lize data
plt.plot(X, y, 'ro')
plt.axis([193, 213, 19, 31])
plt.xlabel('Temperature (F)')
plt.ylabel('Pressure (Atm)')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Tạo mô hình hồi quy tuyến tính
model = LinearRegression().fit(X, y)

# Dự đoán áp suất từ dữ liệu nhiệt độ
y_pred = model.predict(X)

# Vẽ biểu đồ
plt.scatter(X, y, color='red', label='Dữ liệu thực tế')
plt.plot(X, y_pred, color='blue', label='Hồi quy tuyến tính')
plt.xlabel('Nhiệt độ (F)')
plt.ylabel('Áp suất (Atm)')
plt.legend()
plt.title('Hồi quy tuyến tính giữa nhiệt độ và áp suất')
plt.show()
```
