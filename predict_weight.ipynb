```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# height(cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight(kg)
y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
# Visual lize data
plt.scatter(X, y, c = 'red')
plt.axis([140, 190, 45, 70])
plt.title('Visual lize data')
plt.xlabel('Height (cm)')
plt.ylabel('Weight  (kg)')
plt.show()
```

Nhìn vào đồ thị, ta thấy các điểm dữ liệu phân bố gần như một đường thẳng, vậy mô hình Linear Regression nhiều khả năng sẽ cho kết quả tốt.

```python
from sklearn.linear_model import LinearRegression 

# fit the model by Linear Regression
model = LinearRegression() ## fit_intercept = False for calculating the bias
model.fit(X, y)

# Compare two results
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# predict using the fitted model
y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

Chỉ số Mean Squared Error khá nhỏ nghĩa là các giá trị dự đoán khá gần với dữ liệu thực tế.
R-squared gần bầng 1 thể hiện mô hình dự đoán rất tốt dự đoán hoàn hảo các biến mục tiêu.


```python
# Lấy hệ số và điểm cắt
coef = model.coef_[0]
intercept = model.intercept_

# Vẽ đồ thị dữ liệu và đường thẳng hồi quy
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# In hệ số và điểm cắt của đường thẳng
print(f"Coefficient: {coef}")
print(f"Intercept: {intercept}")
```
